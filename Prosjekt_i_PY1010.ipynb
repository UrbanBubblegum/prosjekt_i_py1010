**Del a)**

```python
import pandas as pd    # Importerer Pandas-pakken
import numpy as np    # Importerer Numpy-pakken

fil = "support_uke_24.xlsx"    # Navnet på filen til MORSE med dataloggen for uke 24
data = pd.read_excel(fil)    # Leser Excel-filen i prosjektoppgaven

u_dag = data["Ukedag"].values    # Leser inn verdiene som array fra kolonne 1 i support_uke_24.xlsx
kl_slett = data["Klokkeslett"].values    # Leser inn verdiene som array fra kolonne 2 i support_uke_24.xlsx
varighet = data["Varighet"].values    # Leser inn verdiene som array fra kolonne 3 i support_uke_24.xlsx
score = data["Tilfredshet"].values    # Leser inn verdiene som array fra kolonne 4 i support_uke_24.xlsx
```

**Del b)**

```python
# Summerer antallet ukedager i support_uke_24.xlsx

antall = []    # Lager en tom liste for å lagre antall henvendelser per ukedag i slutten av påfølgende løkke

ukedag = ["Mandag", "Tirsdag", "Onsdag", "Torsdag", "Fredag"]    # Lager liste for 5 ukedager som skal telles

for dag in ukedag:    # Lager en nøstet løkke ("dobbel-løkke") for å telle antall per ukedag i support_uke_24.xlsx
    teller = 0    # teller starter på null for hver dag
    for i in u_dag:    # For hvert element i kolonne 1 (Ukedag) i support_uke_24.xlsx
        if i == dag:    # Hvis elementet i kolonne 1 (Ukedag) i support_uke_24.xlsx er helt lik verdi av gjeldende dag
            teller += 1    # Her telles antallet i løkke 2 
    
    antall.append(teller)    # Legger til antall henvendelser (teller) i listen "antall")

# Visualiserer resultatet i et søylediagram (stolpediagram)

import matplotlib.pyplot as plt    # Importerer matplotlib-pakken

x = ukedag    # Bruker verdi fra listen "ukedag" i del b)
y = antall    # Bruker antall henvendelser fra listen "antall"
farger = ["gold", "darkorange", "lightcoral", "lightgreen", "lightskyblue"]    # Angir fargene som skal brukes på søylediagrammet

plt.close("all")    # Lukker ev. eksterne figurvinduer med plot
plt.bar(x, y, color=farger, width = 0.7)    # Angir at ukedager skal vises på x-aksen, henvendelser på y-aksen, fargelegging, og bredde på søylene
plt.title("Antall henvendelser pr ukedag i uke 24")    # Overskrift på stolpediagrammet
plt.xlabel("Ukedager")    # Tekst på x-aksen
plt.ylabel("Antall henvendelser")    # Tekst på y-aksen

for i in range (0, len(y)):    # Viser antall henvendelser over søylene. Hentet fra læreboken Python for Realfag, 3. utgave, "Eksempel 4.5", side 151)
    plt.text(x[i], y[i]+0.5, str(y[i]))
    
plt.show()    # Koden sørger for at resultatet blir vist
```

**Del c)**

```python
# Numpy-pakken er importert under del a). Her er også array "varighet" (kolonne 3 i support_uke_24.xlsx) lest inn.

a_max = np.argmax(varighet)    # Lager en variabel a_max og bruker funksjonen argmax for å finne den lengste samtaletiden.
print(f"Den lengste samtaletiden i uke 24 er: {varighet[a_max]}")

a_min = np.argmin(varighet)    # Lager en variabel a_min og bruker funksjonen argmin for å finne den korteste samtaletiden. 
print(f"Den korteste samtaletiden i uke 24 er: {varighet[a_min]}")
```

**Del d)**

```python
# Kjøring av denne koden fordrer at del a) er kjørt.

total_tid = []    # Oppretter en tom liste for å fylle samtalevarigheten omregnet i sekunder.

for v in varighet:    # Lager en for-løkke med variabel v til arrayet 'varighet' fra del a)
    tidsdeler = v.split(":")    # Lager en liste som splitter tekst-tallene i tre deler med splitt på kolon
    t, m, s = int(tidsdeler[0]), int(tidsdeler[1]), int(tidsdeler[2])    # Gjør om tekstverdiene til tre heltall: timer, minutter og sekunder
    varighet_sekunder = (t * 3600) + (m * 60) + s    # Lager en liste som konverterer all samtaletid om til sekunder, 1 time = 3600 sekunder, 1 minutt = 60 sekunder
    total_tid.append(varighet_sekunder)    # Legger resultatet for alle samtalene inn i listen total_tid som ble laget først

gjennomsnitt = sum(total_tid) / len(total_tid)    # regner gjennomsnitt av total samtaletid (varighet) i sekunder og deler på antall henvendelser i uke 24

print(f"Gjennomsnittlig samtaletid i uke 24 var {gjennomsnitt:.2f} sekunder")    # Printer resultatet til skjerm i sekunder
```

**Del e)**

```python
# Kjøring av denne kodelinjen fordrer at del a) og del b) er kjørt.
# Dette for å importere numpy og matplotlib

timetall = []    # Oppretter en tom liste for å fylle time-tall i.

for tms in kl_slett:    # Lager en for-løkke med variabel tms til arrayet "kl_slett" fra del a)
    tidsdeler = tms.split(":")    # Lager en liste som splitter tekst-tallene i tre deler med splitt på kolon
    t = int(tidsdeler[0])    # Gjør om tekstverdien for time til et heltall for time
    timer = t    # Lager en liste for alle t-verdiene
    timetall.append(timer)    # Legger resultatet fra listen timer inn i listen timetall som ble laget først

tidsbolk_1 = 0    # Oppretter teller for tidsbolk kl. 08-10 
tidsbolk_2 = 0    # Oppretter teller for tidsbolk kl. 10-12
tidsbolk_3 = 0    # Oppretter teller for tidsbolk kl. 12-14
tidsbolk_4 = 0    # Oppretter teller for tidsbolk kl. 14-16

for t in timetall:    # Lager en for-løkke med if og elif for variabel t i listen timetall
    if 8 <= t < 10:    # Større enn eller lik 8 men mindre enn 10
        tidsbolk_1 += 1    # Summerer opp antall score fra support_uke_24.xlsx
    elif 10 <= t < 12:    # Større enn eller lik 10 men mindre enn 12
        tidsbolk_2 += 1    # Summerer opp antall score fra support_uke_24.xlsx
    elif 12 <= t < 14:    # Større enn eller lik 12 men mindre enn 14
        tidsbolk_3 += 1    # Summerer opp antall score fra support_uke_24.xlsx
    elif 14 <= t < 16:    # Større enn eller lik 14 men mindre enn 16
        tidsbolk_4 += 1    # Summerer opp antall score fra support_uke_24.xlsx

antall = [tidsbolk_1, tidsbolk_2, tidsbolk_3, tidsbolk_4]    # Lager en liste som henter samlet antall for hver tidsbolk 1, 2, 3 og 4
         
henvendelser  = np.array(antall)    # Lager en variabel som er array av antall 

# tidsbolker (under) er her labels. Jeg hentet inspirasjon til denne fra læreboken Python for Realfag, 3. utgave, "Listing 4.6: prog_kakediagram.py", side 153)
tidsbolker = ["kl. 08-10: " + str(antall[0]), "kl. 10-12: " + str(antall[1]), "kl. 12-14: " + str(antall[2]), "kl. 14-16: " + str(antall[3])]

farger = ["gold", "lightcoral", "lightgreen", "lightskyblue"]    # Gjenbruker farger fra del b)

plt.close("all")    # Lukker ev. eksterne figurvinduer med plot
plt.pie(henvendelser, labels = tidsbolker, startangle = 135, autopct = "%1.1f%%", colors=farger)    # data som skal vises i plottet
plt.title("Totale antall henvendelser i de ulike tidsrommene i uke 24")    # Forklarende tittel til sektordiagrammet
plt.show()    # Viser plottet      
```

**Del f)**

```python
# Kjøring av denne kodelinjen fordrer at del a) er kjørt.

negativ = 0    # Oppretter teller for negativ score
noytral = 0    # Oppretter teller for nøytral score
positiv = 0    # Oppretter teller for positiv score

for feedback in score:    # Lager en for-løkke med if og elif for variabel feedback til array "score" fra del a)
    if 1 <=  feedback <= 6:    # Fra og med 1 til og med 6
        negativ += 1    # Summerer opp antall i listen for alle 217 rader
    elif 7 <=  feedback <= 8:   # Fra og med 7 til og med 8
        noytral += 1    # Summerer opp antall i listen for alle 217 rader
    elif 9 <= feedback <= 10:    # Fra og med 9 til og med 10
        positiv += 1    # Summerer opp antall i listen for alle 217 rader
           
total = negativ + noytral + positiv    # Oppretter liste som summerer sammen antallet negative, nøytrale og positive score

# NPS = % positive kunder - % negative kunder (hentet fra oppgaveteksten til del f)

nps = (positiv * 100) / total - (negativ * 100) / total

print(f"Supportavdelingens NPS for uke 24 er {nps:.2f} %. Avrundet til nærmeste hele prosenttall blir det {nps:.0f} %.")
```
